Let's replace the columns 'Not Available', 'unknown' with 'nan' (null value)

```python
df = pd.read_csv('orders.csv',na_values = ['Not Available','unknown'])
```

```python
df["Ship Mode"].unique()
```

Let's rename the columns in our dataset:

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

Let's derive some new columns in our dataset

```python
df["discount"]= df["list_price"]*df["discount_percent"]*0.01
df["selling_price"] = df["list_price"] - df["discount"]
df["profit"] = df["selling_price"]-df["cost_price"]
```

```python
df.head()
```

Let's take a look at the datatypes of all columns:

```python
df.dtypes
```

Here we notice that the order_date column has the datatype object, however when we use the SQL server to analyze the data we would want the order_date column to have datatype datetime instead. So let's change the datatype of the order_date column to datetime instead

```python
df["order_date"]=pd.to_datetime(df["order_date"],format="%Y-%m-%d")
```

Let's again take a look at the datatypes of all columns:

```python
df.dtypes
```

The columns list_price and discount_percent seem pretty unnecessary since we have already created new columns: selling_price and discount. I will be dropping these 2 columns.

```python
df.drop(columns=["list_price","discount_percent"],inplace=True)
```
